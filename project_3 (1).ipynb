```python
def f(x):
    if x<10:
        return 0
    elif x<20:
        return 1.0
    elif x<30:
        return 2.0
    elif x<40:
        return 3.0
    elif x<50:
        return 4.0
    else:
        return 5.0
           
```

```python
x=data[data['rating'].isnull()]
```

```python
x
```

```python
y=data[~data['rating'].isnull()]
```

```python
y
```

```python
x['rating']=x['order_frequency'].apply(f)
```

```python
x
```

```python
final=pd.concat((x,y),ignore_index=True)
```

```python
final
```

```python
final.isnull().sum()
```

```python
delivery = delivered_data.head(20).copy()

delivery['order_frequency'] = delivery['order_frequency'].astype(float)
delivery['dish_name'] = delivery['dish_name'].astype(str)

import matplotlib.pyplot as plt

plt.figure(figsize=(10,5),facecolor='lightgreen')
plt.bar(delivery['dish_name'], delivery['order_frequency'], color='blue',width=0.5)

ax=plt.gca()
ax.set_facecolor('pink')


plt.ylabel("Order Frequency")
plt.title("Order Frequency per Dish")
plt.show()
```

```python

plt.figure(figsize=(10,5),facecolor='aqua')
ax=plt.gca()
ax.set_facecolor('floralwhite')
plt.scatter(delivery['age'],delivery['dish_name'],
           marker='H',edgecolor='k',facecolor='red',s=100,alpha=0.5,label='age')
plt.xlabel('age')
plt.ylabel('dish')
plt.title('dish ordered by different age groups')
plt.legend()
```

```python
avgdish=delivery.groupby('dish_name')['rating'].mean()

plt.figure(figsize=(7,7),facecolor='pink')
c=['red','green','yellow','indigo','orange']
plt.pie(avgdish,colors=c,shadow=True,labels=avgdish.index,
        wedgeprops={'edgecolor':'k'})
```
